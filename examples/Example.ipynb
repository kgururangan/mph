```python
import numpy as np
from matplotlib import pyplot as plt
from mph.mph import MPH
from mph.utilities import plot_spectrum
```

```python
nmol = 20
vibmax = 5
Ef = 14000
omega = 1400
J = np.zeros(nmol)
```

```python
Jval = [100, 500, 700, 1000]
lambda_f = 2.0
for val in Jval:
    J[1] = val
    H = MPH(nmol, vibmax, omega, Ef, J, lambda_f)
    H.kernel()
    H.get_abs_spectrum(gamma=250)
    plt.plot(H.photon_energy/H.omega, H.absorbance)
```

```python
J[1] = 700

lambda_f = [0.0, 0.5, 1.0, 1.5, 2.0]

for val in lambda_f:
    H = MPH(nmol, vibmax, omega, Ef, J, val)
    H.kernel()
    H.get_abs_spectrum(gamma=250)
    plt.plot(H.photon_energy/H.omega, H.absorbance)
```

```python
J[1] = 700
lambda_f = 0.5

H = MPH(nmol, vibmax, omega, Ef, J, lambda_f)
H.kernel()
H.get_abs_spectrum(gamma=250)
plot_spectrum(H)
```
